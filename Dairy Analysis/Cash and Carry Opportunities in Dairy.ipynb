```python
import json
import matplotlib.pyplot as plt
from datetime import datetime

# Hard-coded values that can be adjusted as needed
storage_costs_per_month = 50  # Example storage cost per month
handling_cost = 100  # Example handling cost
annual_financing_rate = 0.05  # Example annual financing rate
nzd_to_usd = 0.62
eur_to_usd = 1.085

# Load the JSON data
with open('futures_data.json', 'r') as file:
    data = json.load(file)

# Extract spot prices from the provided JSON and convert if necessary
spot_prices = data["spot_prices"]
spot_prices_converted = {
    "NDFM_CME": spot_prices["CME"]["NDFM"] * 2204.62,  # Convert from $/lb to $/MT
    "Butter_CME": spot_prices["CME"]["Butter"] * 2204.62,  # Convert from $/lb to $/MT
    "Cheese_CME": spot_prices["CME"]["Cheese"] * 2204.62,  # Convert from $/lb to $/MT
    "Block_Cheese_CME": spot_prices["CME"]["Block Cheese"] * 2204.62,  # Convert from $/lb to $/MT
    "Whey_CME": spot_prices["CME"]["Whey"] * 2204.62,  # Convert from $/lb to $/MT
    "Class_III_Milk_CME": spot_prices["CME"]["Class III Milk"] * 22.0462,  # Convert from $/cwt to $/MT
    "Class_IV_Milk_CME": spot_prices["CME"]["Class IV Milk"] * 22.0462,  # Convert from $/cwt to $/MT
    "SMP_EEX": spot_prices["EEX"]["SMP"] * eur_to_usd,  # Convert from /MT to $/MT
    "Butter_EEX": spot_prices["EEX"]["Butter"] * eur_to_usd,  # Convert from /MT to $/MT
    "Whey_EEX": spot_prices["EEX"]["Whey"] * eur_to_usd,  # Convert from /MT to $/MT
    "Liquid_Milk_EEX": spot_prices["EEX"]["Liquid Milk"] * 10.85,  # Convert from /100 kg to $/MT using conversion factor
    "SMP_SGX": spot_prices["SGX"]["SMP"],  # Already in $/MT
    "WMP_SGX": spot_prices["SGX"]["WMP"],  # Already in $/MT
    "AMF_SGX": spot_prices["SGX"]["AMF"],  # Already in $/MT
    "Butter_SGX": spot_prices["SGX"]["Butter"],  # Already in $/MT
    "Milk_Price_SGX": spot_prices["SGX"]["Milk Price"] / nzd_to_usd * 1000,  # Convert from NZD/kgMS to USD/MT
    "SMP_GDT": spot_prices["GDT"]["SMP"],  # Already in $/MT
    "WMP_GDT": spot_prices["GDT"]["WMP"],  # Already in $/MT
    "AMF_GDT": spot_prices["GDT"]["AMF"]  # Already in $/MT
}

# Convert futures prices for specific products to $/MT if needed
def convert_futures_price(product, price):
    if "Class III Milk" in product or "Class IV Milk" in product:
        return price * 22.0462  # Convert from $/cwt to $/MT
    elif "Liquid Milk" in product:
        return price * 10.85  # Convert from /100 kg to $/MT
    elif "Milk Price" in product:
        return price / nzd_to_usd * 1000  # Convert from NZD/kgMS to USD/MT
    return price  # No conversion needed

def analyze_and_plot(futures_data, spot_price, storage_cost_per_month, handling_cost, annual_financing_rate, product_name, unit):
    print(f"\nAnalyzing {product_name} with spot price {spot_price}")
    
    # Function to calculate carry cost
    def calculate_carry_cost(months):
        monthly_financing_cost = (annual_financing_rate * spot_price) / 12
        total_carry_cost = (storage_cost_per_month * months) + handling_cost + (monthly_financing_cost * months)
        return total_carry_cost

    # Initialize variables to track the best month and max profit
    best_month = None
    max_profit = -float('inf')

    # Lists to store data for plotting
    months_list = []
    futures_prices = []
    profits = []

    # Function to get number of months to expiration
    def months_to_expiration(future_month):
        future_date = datetime.strptime(future_month, "%b-%y")
        current_date = datetime.now().replace(day=1)
        months = (future_date.year - current_date.year) * 12 + (future_date.month - current_date.month)
        return months

    # Iterate through all futures contracts
    for future in futures_data:
        if isinstance(future, dict):
            expiration_month = future["month"]
            futures_price = future["settle"]
            futures_price_converted = convert_futures_price(product_name, futures_price)

            # Calculate the number of months from now until the expiration date
            months = months_to_expiration(expiration_month)
            if months < 0:
                continue  # Skip expired contracts

            # Calculate the carry cost
            total_carry_cost = calculate_carry_cost(months)

            # Calculate the potential profit for both regular and inverse cash-and-carry
            potential_profit = futures_price_converted - (spot_price + total_carry_cost)
            inverse_potential_profit = (spot_price - total_carry_cost) - futures_price_converted

            # Debug: Print intermediate values
            print(f"Month: {expiration_month}, Futures Price: {futures_price_converted}, Total Carry Cost: {total_carry_cost}, Potential Profit: {potential_profit}, Inverse Potential Profit: {inverse_potential_profit}")

            # Append data for plotting
            months_list.append(expiration_month)
            futures_prices.append(futures_price_converted)
            profits.append(max(potential_profit, inverse_potential_profit))

            # Check if this is the best month so far
            if max(potential_profit, inverse_potential_profit) > max_profit:
                max_profit = max(potential_profit, inverse_potential_profit)
                best_month = expiration_month

    # Print the results
    if max_profit > 0:
        print(f"The best month for the Cash-and-Carry trade for {product_name} is: {best_month}")
        print(f"Maximum potential profit: ${max_profit:.2f} per {unit}")
    else:
        print(f"No arbitrage opportunity found for {product_name}")

    # Plotting the data with twin y-axes
    fig, ax1 = plt.subplots(figsize=(10, 6))

    ax1.set_xlabel('Month')
    ax1.set_ylabel(f'Price (USD per {unit})', color='tab:blue')
    ax1.plot(months_list, futures_prices, label='Futures Prices', marker='o', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()  # Instantiate a second y-axis that shares the same x-axis
    ax2.set_ylabel(f'Profit (USD per {unit})', color='tab:orange')
    ax2.plot(months_list, profits, label='Potential Profit', marker='x', color='tab:orange')
    ax2.tick_params(axis='y', labelcolor='tab:orange')
    ax2.axhline(0, color='red', linestyle='--', linewidth=1)

    # Annotate the best month if there is a positive profit
    if best_month and max_profit > 0:
        ax2.annotate(f'Max Profit: ${max_profit:.2f} per {unit}', 
                     xy=(best_month, max_profit), 
                     xytext=(best_month, max_profit + 0.5),
                     arrowprops=dict(facecolor='black', arrowstyle='->'),
                     fontsize=10)

    fig.tight_layout()  # Otherwise the right y-label is slightly clipped
    plt.title(f'Futures Prices and Potential Profits for {product_name}')
    plt.xticks(rotation=45)
    plt.show()

# Analyzing and plotting for multiple products and exchanges
for exchange, products in data["futures"].items():
    for product_name, futures_data in products.items():
        # Create the correct spot price key
        formatted_product_name = product_name.replace(' ', '_').replace('-', '_')
        spot_price_key = f"{formatted_product_name}_{exchange}"
        spot_price = spot_prices_converted.get(spot_price_key)
        if spot_price is None:
            print(f"No spot price found for {product_name} on {exchange}")
            continue
        print(f"Product: {product_name}, Exchange: {exchange}, Spot Price Key: {spot_price_key}, Spot Price: {spot_price}")
        analyze_and_plot(futures_data, spot_price, storage_costs_per_month, handling_cost, annual_financing_rate, f'{exchange} {product_name}', 'MT')
```
